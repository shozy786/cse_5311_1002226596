```python
import random
def qsort(arr, start, end):
    if start >= end:
        return arr
    
    pivot = random.choice([start + i for i in range(end-start+1)]) # Random Pivot everytime
    
    arr[pivot], arr[end] = arr[end], arr[pivot]
    pivot = end
    left = start
    right = end - 1
    
    while left <= right:
        if arr[left] > arr[pivot] and arr[right] < arr[pivot]:
            arr[left], arr[right] = arr[right], arr[left]
            left += 1
            right -= 1
        elif arr[left] <= arr[pivot]:
            left += 1
        elif arr[right] >= arr[pivot]:
            right -= 1
   
    arr[left], arr[end] = arr[end], arr[left]
    qsort(arr, start, left - 1)
    qsort(arr, left+1, end)
    return arr

def quick_sort(A):
    qsort(A, 0, len(A)-1)
```

```python
import numpy as np
from time import perf_counter
from copy import copy
N = [(i+1000)*10 for i in range(1000)]
arrays = {}
for n in N:
    arrays.update({n: np.random.randint(0, 1000, n).tolist()})
```

```python
# Random Pivot
random_times = []
for n in N:
    arr = arrays[n]
    t1 = perf_counter()
    quick_sort(arr)
    random_times.append(perf_counter() - t1)
```

```python
import matplotlib.pyplot as plt

plt.scatter(N, random_times, label = "Random Time", color = 'green')
plt.title("Quick Sort with random pivot Time")
plt.xlabel("n - Size of Array")
plt.ylabel("Time (s)")
plt.legend(loc = "upper left")
plt.show()
```
