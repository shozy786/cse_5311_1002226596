# Results Verification

```python
def is_sorted(arr):
    for i in range(len(arr)-1):
        if arr[i] > arr[i+1]:
            return False
    return True
```

```python
sample = np.random.randint(0, 1000, size=1000)

# 1. Insertion Sort
sorted_arr = insertion_sort(sample)
if is_sorted(sorted_arr):
    print("Insertion Sort code is working.")
# 2. Selection Sort
sorted_arr = selection_sort(sample)
if is_sorted(sorted_arr):
    print("Selection Sort code is working.")
# 3. Bubble Sort
sorted_arr = bubble_sort(sample)
if is_sorted(sorted_arr):
    print("Bubble Sort code is working.")
```

```python
# Let's define our test array sizes
N = [5, 10, 20, 50, 100, 150, 200, 250, 300, 350, 400, 450, 500, 550, 600, 650, 700, 750, 800, 850, 900, 1000, 1050, 1100, 1150, 1200, 1250, 1300, 1350, 1400, 1450, 1500, 1550, 1600, 1650, 1700, 1750, 1800, 1850, 1900, 1950, 2000]
```

```python
insertion_benchs = []
selection_benchs = []
bubble_benchs = []

for n in N:
    setup = f"""
from insertion import insertion_sort
from selection import selection_sort
from bubble import bubble_sort
import numpy as np
arr = np.random.randint(0, 1000, size={n})
    """
    i = timeit(stmt = "insertion_sort(arr)", setup=setup,number=100)
    s = timeit(stmt = "selection_sort(arr)", setup=setup,number=100)
    b = timeit(stmt = "bubble_sort(arr)", setup=setup,number=100)
    insertion_benchs += [i]
    selection_benchs += [s]
    bubble_benchs += [b]    
```

```python
plt.plot(N, insertion_benchs, label = "Insertion Sort")
plt.plot(N, selection_benchs, label = "Selection Sort")
plt.plot(N, bubble_benchs, label = "Bubble Sort")
plt.xlabel("Number of Iterations")
plt.ylabel("Time (s)")
plt.legend(loc = "upper left")
plt.show()
```
